```python
cumulative_infiltration_matrix = []
cumulative_val = 0.0
for i in range (0,len(raw_data['vTopM'])):
    to_add = raw_data['vTopM'][i]/12
    cumulative_val = cumulative_val + to_add
    cumulative_infiltration_matrix.append(cumulative_val)
plt.plot(cumulative_infiltration_matrix)

cumulative_precip = []
cumulative_val = 0.0
for i in range (0,len(raw_data['rTopT'])):
    to_add = raw_data['rTopT'][i]/12
    cumulative_val = cumulative_val + to_add
    cumulative_precip.append(cumulative_val)

plt.plot(cumulative_precip)

    
```

```python
cumulative_infiltration_fracture = []
cumulative_val = 0.0
for i in range (0,len(raw_data['vTopF'])):
    to_add = raw_data['vTopF'][i]/12
    cumulative_val = cumulative_val + to_add
    cumulative_infiltration_fracture.append(cumulative_val)
plt.plot(cumulative_infiltration_fracture)
```

```python
cumulative_infiltration = []
cumulative_runoff = []

for item1,item2,item3 in zip(cumulative_infiltration_matrix, cumulative_infiltration_fracture, cumulative_precip):
    cumulative_infiltration.append(-item1+-item2)
    cumulative_runoff.append(-1*(-item1+-item2 + item3))
plt.plot(hours_HYDRUS, cumulative_infiltration, linestyle='dashed', label='HYDRUS infiltration', color='blue')
plt.plot(hours_HYDRUS, cumulative_runoff, linestyle='dashed', label='HYDRUS runoff', color='orange')
plt.plot(hours_HYDRUS, -1*raw_data['sum(vBotT)'],linestyle='dashed',label='HYDRUS recharge', color='purple')
plt.plot(hours_HYDRUS, raw_data['sum(vRoot)'],linestyle='dashed',label='HYDRUS AET',color='green')


plt.plot(hours_LGARTO, np.cumsum(100*(data_c_vars['infiltration'] + data_c_vars['infiltration_frac'])),label='LGARTO infiltration', color='blue')
plt.plot(hours_LGARTO, np.cumsum(100*(data_c_vars['surface_runoff_frac'])),label='LGARTO runoff', color='orange')
plt.plot(hours_LGARTO, np.cumsum(100*(data_c_vars['percolation'] + data_c_vars['percolation_frac'])),label='LGARTO recharge', color='purple')
plt.plot(hours_LGARTO, np.cumsum(100*data_c_vars['actual_evapotranspiration']),label='LGARTO AET',color='green')


plt.xlabel("simulation hour")
plt.ylabel("cumulative flux (cm)")

# Place legend below plot with two columns
plt.legend(
    loc='upper center',
    bbox_to_anchor=(0.5, -0.15),
    ncol=2,
    frameon=False
)

plt.tight_layout()  # Adjust layout to make room for legend
plt.savefig(sim_name+'.pdf', bbox_inches='tight')
```
